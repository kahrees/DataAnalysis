### Splitting for the Specified Columns

```python

gd['Event Description'] = pd.Series(gd['Event Description'], dtype="string")
gd['Geographic Areas'] = pd.Series(gd['Geographic Areas'], dtype="string")
gd['Respondent'] = pd.Series(gd['Respondent'], dtype="string")
gd['NERC Region'] = pd.Series(gd['NERC Region'], dtype="string")

gd.info()
```

```python
#Standardize Event Description Before Split

#import re

gd['Event Description'] = gd['Event Description'].str.replace('Suspected Physical Attack','Physical Attack - Suspected')
gd['Event Description'] = gd['Event Description'].str.replace('Physical Attack; Vandalism','Physical Attack - Vandalism')
gd['Event Description'] = gd['Event Description'].str.replace('Hurricane {%}','Severe Weather - Hurricane {%}')
```

```python

# Splitting the Event Description column by the '-' delimiter
event_split = gd['Event Description'].str.split('-', n = 16, expand=True)

# Renaming columns for better clarity
event_split.columns = ['Event Category', 'Event Type']

# Displaying the split address DataFrame
print("\nNew DataFrame:")
print(event_split)
```
